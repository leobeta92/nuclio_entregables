```python
!pip install matplotlib

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import plotly.express as px

np.random.seed(777)

CWD = os.getcwd()
HOST_DATA_PATH = os.path.join(CWD, 'airbnb_madrid_host.parquet')
CONDITIONS_DATA_PATH = os.path.join(CWD, 'airbnb_madrid_conditions.parquet')
LOCATION_DATA_PATH = os.path.join(CWD, 'airbnb_madrid_location.parquet')
PROPERTY_DATA_PATH = os.path.join(CWD, 'airbnb_madrid_property.parquet')
REVIEWS_DATA_PATH = os.path.join(CWD, 'airbnb_madrid_reviews.parquet')

# Load datasets
host_df = pd.read_parquet(HOST_DATA_PATH)
conditions_df = pd.read_parquet(CONDITIONS_DATA_PATH)
property_df = pd.read_parquet(PROPERTY_DATA_PATH)
reviews_df = pd.read_parquet(REVIEWS_DATA_PATH)
location_df = pd.read_parquet(LOCATION_DATA_PATH)

# Merge datasets

airbnb_df = host_df.merge(conditions_df, on= 'id').merge(location_df, on = 'id').merge(property_df, on = 'id').merge(reviews_df, on = 'id')
```

```python
airbnb_df.shape
```

```python
# property_types = ['apartment','house','condominium','loft','serviced_apartment']
property_types = ['apartment']

airbnb_analisis = airbnb_df.loc[airbnb_df['property_type'].isin(property_types)]
```

```python
airbnb_price_count = airbnb_analisis['price']
airbnb_price_count

price_hist = px.histogram(
    data_frame = airbnb_price_count,
    x = 'price',
    nbins = 200,
)

price_hist.update_traces(
    marker=dict(
        line=dict(color="black", width=1)   # outline
    )
)
```

```python
print(airbnb_analisis['price'].agg(['count','median','mean','min','max','std']).round(2))
print('-'*15)
print('Mode:', airbnb_analisis['price'].mode())
```

Viendo por barrio, sí se parece diferente cuando solo vemos los apartamentos.

```python
airbnb_analisis['economico'] = airbnb_analisis['price'] < airbnb_analisis['price'].quantile(0.25)
airbnb_analisis['standard'] = (airbnb_analisis['price'] > airbnb_analisis['price'].quantile(0.25)) & (airbnb_analisis['price'] < airbnb_analisis['price'].quantile(0.75))
airbnb_analisis['lux'] = (airbnb_analisis['price'] > airbnb_analisis['price'].quantile(0.75)) & (airbnb_analisis['price'] < airbnb_analisis['price'].quantile(0.90))
airbnb_analisis['ultra_lux'] = airbnb_analisis['price'] > airbnb_analisis['price'].quantile(0.90)
```

```python
def price_category(row):
    if row['economico']:
        return '1- economico'
    elif row['standard']:
        return '2- standard'
    elif row['lux']:
        return '3- lux'
    elif row['ultra_lux']:
        return '4- ultra_lux'
```

```python
airbnb_analisis['price_category'] = airbnb_analisis.apply(price_category, axis = 1)
```

```python
airbnb_price_count = airbnb_analisis.groupby('price_category', as_index= False)['id'].count().reset_index()

plt.figure(figsize=(12, 6))

# TODO: boxplot count vs price category
# ax = sns.boxplot(
#     data=airbnb_df,
#     y= 'price',
#     x="price_category",
#     legend=False
#     )

px.bar(
    airbnb_price_count,
    x = 'price_category',
    y = 'id',
)
```
