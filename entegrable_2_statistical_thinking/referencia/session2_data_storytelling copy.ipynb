# 1. Definición del Problema

Como es común en todos los proyectos, partimos de un problema inicial escueto
y sencillo:

<div style="text-align:center; color:red; font-size:larger; font-weight:bold; border: 2px solid red; padding: 10px;">
    (1) ¿De qué depende el valor de un futbolista?
</div>

El primer paso en el que debemos trabajar para construir un buen
`data storytelling` es en dar con una definición de un problema que cumpla con
los siguientes requisitos:

- es ``concreto``: no hay términos ambiguos.
- el ``alcance`` está definido: entendemos dónde empieza y dónde acaba el
ejercicio.
- tiene ``valor`` para el negocio: si damos con una respuesta, el negocio se
verá beneficiado.
- la respuesta es ``accionable``: la respuesta puede traducirse en acciones
reales que estén al alcance de la compañía.

<div style="text-align:center; color:red; font-size:larger; font-weight:bold; border: 2px solid red; padding: 10px;">
    Discutir si la definición inicial del problema cumple con los 4 requisitos
    de un problema bien enunciado.
</div>

- ``concreto`` $\rightarrow$ ¿qué entendemos por ``valor`` de un futbolista?
- ``alcance`` $\rightarrow$ ¿qué ``jugadores`` analizamos? ¿en qué ``periodo temporal`` van a ser analizados?
- ``valor`` $\rightarrow$ ¿qué ``valor`` tiene para negocio responder a este problema?
- ``accionable`` $\rightarrow$ ¿cómo puede la compañía ``accionar`` la respuesta de este problema?

Si tenemos en cuenta estos 4 aspectos para formular correctamente un problema,
podríamos convertir el problema (1) en (2):

<div style="text-align:center; color:red; font-size:larger; font-weight:bold; border: 2px solid red; padding: 10px;">
    (2) ¿Qué variables muestran mayores diferencias en el valor de mercado que
    tienen el top 10% de los futbolistas de la temporada 2023-2024 reconocidos
    por la FIFA con respecto a un jugador promedio?
</div>

- ``concreto`` $\rightarrow$ entendemos por ``valor`` de un futbolista como el valor de mercado estrictamente, cuánto estaría el mercado dispuesto a pagar por hacerse por el jugador. Dejamos a un lado aspectos como pueden ser beneficios por derechos de imagen o estrategias para abrir mercado en ventas de merchandising en nuevos paises, por ejemplo. En muchos deportes, contar con una estrella en tu equipo de un pais en concreto puede ayudar al departamento de marketing a abrir nuevos mercados.
- ``alcance`` $\rightarrow$ los ``jugadores`` que vamos a analizar son aquellos reconocidos por la FIFA en la temporada 2023-2024 y querremos comparar el top 10% de los jugadores con mayor valor con un jugador promedio.
- ``valor`` $\rightarrow$ si conseguimos entender en qué variables debemos fijarnos para determinar el ``valor`` de un futbolista, el club podrá mejorar su estrategia de fichajes reduciendo el riesgo y siendo más competitivo. En primer lugar en términos económicos, este problema puede ayudar a la salud financiera del club y si se mejora la calidad de los fichajes también en el aspecto deportivo.
- ``accionable`` $\rightarrow$ tras encontrar las variables que más pueden influir en el valor de un futbolista, el club podrá fijarse solo en las variables relevantes para tomar mejores decisiones en materia de ``fichajes``. El club puede encontrar jugadores con buenas cualidades que estén por debajo del precio que deberían y evitar invertir de más en jugadores que estén por encima de mercado.

# 2. Analysis Roadmap

Una vez hemos formulado correctamente el problema que queremos resolver, damos
paso a la face de análisis. Pero no debemos lanzarnos directamente contra los
datos. Es recomendable plantear un ``analysis roadmap`` atendiendo a la
siguiente estructura:

Los beneficios de seguir este enfoque son varios:
- ganamos más foco y eficiencia en el desarrollo de análisis evitando el caos
inicial de análisis desordenados
- ayuda al equipo a mantenerse alineado en los objetivos de análisis que vamos
a cubrir (por lo menos inicialmente)
- es más fácil dividir las tareas independientes entre los miembros del equipo
- ganamos estructura que nos ayudará a montar con más claridad el data
storytelling final que decidamos

Un buen analysis roadmap debe cumplir los siguientes requisitos:
- las hipótesis y las preguntas deben ser ``exhaustivas``
- las hipótesis deben ser ``independientes``
- responder a cada pregunta debe ayudarnos a dar ``respuesta al problema planteado``

## Hipótesis 3: capacidad ofensiva

### questions 3.1 - 3.10

Cuando analizamos muchos datos, una cuestión que nos surge cuando tenemos que
visualizar muchas variables es intentar unificar en un único plot varios
gráficos. Plotly Express nos ofrece la virtud de poder generar gráficos
interactivos de una forma muy sencilla, pero personalizar gráficos no lo es
tanto. No obstante, podemos aprovechar el argumento ``facet_col`` o
``facet_row`` para agrupar visualizaciones con un poquito de imaginación
manipulando los datos originales:

```python
# TODO: definir variables a agrupar

fts_ofensivas1 = ["regate", "posicion_de_ataque", "cabezazo", "potencia_de_tiro", "definicion"]
fts_ofensivas2 = ["tiros_de_larga_distancia", "curva", "precision_en_tiro_libre", "penaltis", "voleas"]
fts_ofensivas3 = fts_ofensivas1 + fts_ofensivas2

fts1 = fp_mod[fts_ofensivas1 + ['is_top_10']].melt(id_vars= 'is_top_10')
fts2 = fp_mod[fts_ofensivas2 + ['is_top_10']].melt(id_vars= 'is_top_10')
fts3 = fp_mod[fts_ofensivas3 + ['is_top_10']].melt(id_vars= 'is_top_10')
```

```python
# TODO: boxplot para el grupo 1

fig1 = px.box(
    data_frame= fts1,
    x = 'is_top_10',
    y = 'value',
    color = 'is_top_10',
    points = 'all',
    facet_col= 'variable'
)
fig1.show()

# TODO: boxplot para el grupo 2

fig2 = px.box(
    data_frame= fts2,
    x = 'is_top_10',
    y = 'value',
    color = 'is_top_10',
    points = 'all',
    facet_col= 'variable'
)
fig2.show()
```

```python
fig3 = px.box(
    data_frame= fts3,
    x = 'is_top_10',
    y = 'value',
    color = 'is_top_10',
    points = 'all',
    facet_col= 'variable'
)
fig3.show()
```
