## Observciones
A la primera vista, parece que hay distribuciónes muy grandes entre los precios, con algunos muy, muy altos. La histograma "skews right", que significa que la media es más grande que la mediana, y más grande que el mode. Entonces, si trato de analisar el dataset completo, usando le precio medio no va aplicar a la gran mayoría de propiedades.

No se sabe que tipo de portafolio tiene el cliente, pero se puede explorar formas de optimizar "listings" para las propiedades de bajo, media, y alta categoria. O, como dirían a los clientes: economico, standard, y lux. Tal vez dividiendo los "listings" por quantil. 

Quiero ver como son los quantiles 0.25, 0.5, y 0.75 por barrio. A ver si puedo aplicar la definición "economico", "standard", y "lux".

### Hipotesis 1
Las propiedades en los diferentes niveles ofrecen servicios distintos. Una propiedad de menor precio ofrece menos que una propiedad de un precio más alta. Entonces, para poder subir una propiedad hasta la próxima categoria, hay que buscar a proveer amenidades que tienen esas propiedades.

```python
airbnb_df['price'].quantile([0.25, 0.5, 0.75, 0.90])
```

```python
px.box(
    data_frame= airbnb_df,
    x = 'price',
    y = 'neighbourhood_group',
    range_y= 0,
    # points = 'all',
    height = 1000,
    width = 1000
)
```

```python
pd.pivot_table(airbnb_df.groupby('neighbourhood_group')['price'].quantile([0.25, 0.5, 0.75, 0.95]).reset_index(), values = 'price', index= ['neighbourhood_group'], columns = 'level_1')
```

Sí hay variaciones grandes entre barrios, donde el mediano de algunas es más grande que el quantil(.75) de otras. Por ejemplo, Retiro comparado a Latina. Pero, parece que si puedo usar los quantiles 0.25, 0.25-0.75, 0.75-0.90, y >0.9 para definir 4 categorías.  

```python
airbnb_df['economico'] = airbnb_df['price'] < airbnb_df['price'].quantile(0.25)
airbnb_df['standard'] = (airbnb_df['price'] > airbnb_df['price'].quantile(0.25)) & (airbnb_df['price'] < airbnb_df['price'].quantile(0.75))
airbnb_df['lux'] = (airbnb_df['price'] > airbnb_df['price'].quantile(0.75)) & (airbnb_df['price'] < airbnb_df['price'].quantile(0.90))
airbnb_df['ultra_lux'] = airbnb_df['price'] > airbnb_df['price'].quantile(0.90)
```

```python
def price_category(row):
    if row['economico']:
        return '1- economico'
    elif row['standard']:
        return '2- standard'
    elif row['lux']:
        return '3- lux'
    elif row['ultra_lux']:
        return '4- ultra_lux'
```

```python
airbnb_df['price_category'] = airbnb_df.apply(price_category, axis = 1)
```
